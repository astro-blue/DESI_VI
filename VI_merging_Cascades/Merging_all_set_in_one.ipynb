```python
import pandas as pd
import numpy as np
```

```python
merged_dir = '/global/u2/t/tlan/projects/VI_files/Cascades/BGS/output/'
merged_files=['desi-vi_SV_subsample_BGS_sel0_all_210513.csv','desi-vi_SV_subsample_BGS_sel1_all_210513.csv']
set_name = ['sel0','sel1']
combined_file = merged_dir+'desi-vi_SV_cascades_combination_BGS_all_210521.csv'
vimerged = pd.read_csv(merged_dir + merged_files[0], delimiter = ",", engine='python',keep_default_na=False)
vimerged['setting_name']=set_name[0]

# Read in the rest of the files and append them to vi
for i in range(1,len(merged_files)):
    print(merged_files[i])
    vi2 = pd.read_csv(merged_dir + merged_files[i], delimiter = ",", engine='python',keep_default_na=False)
    vi2['setting_name']=set_name[i]
    vimerged = vimerged.append(vi2, ignore_index=True)

vimerged.to_csv(combined_file,index=False)
```
