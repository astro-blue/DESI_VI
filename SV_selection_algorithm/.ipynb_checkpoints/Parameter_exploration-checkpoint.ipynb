```python
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
import pandas as pd
import my_wise_utility_v2 as my_wise_ut
```

```python
dir = '/Users/tlan/Dropbox/Astro_Research/Projects_plots_notes/2020_DESI_visual_inspect/SV1/Truth_tables/'
filename = 'desi-vi_ELG_tile80608_nightdeep_merged_all_210208_ADDING_many_info.csv'
data1 = pd.read_csv(dir+filename,keep_default_na=False)
filename = 'desi-vi_ELG_tile80606_nightdeep_merged_all_210218_ADDING_many_info.csv'
data2 = pd.read_csv(dir+filename,keep_default_na=False)
frame = [data1,data2]

data = pd.concat(frame)
data = data[(data['ZWARN']==0)]
#very_good_data = data[(data['best_quality']==4) & (abs(data['best_z']-data['Redrock_z'])/data['best_z']<0.033)]
good_data = data[(data['best_quality']<=4) & (data['best_quality']>=2.5) & (abs(data['best_z']-data['Redrock_z'])/(1+data['best_z'])<0.0033)]
len(data)
print(len(data.loc[data['DELTACHI2']>100]))
print(len(good_data.loc[good_data['DELTACHI2']>100]))
print(len(good_data.loc[good_data['DELTACHI2']>10])/len(data.loc[data['DELTACHI2']>10]))
```

```python
plt.hist(np.log10(data['DELTACHI2']),bins=20,range=(0,5))
plt.hist(np.log10(good_data['DELTACHI2']),bins=20,range=(0,5))
```

1st thing to explore: the redrock redshift success rate vs DELTACHI2

```python
log_delta_bin = my_wise_ut.my_bin_maker(1,4,20)
```

```python
x_all = np.log10(data['DELTACHI2'])
print(len(x_all))
x_good = np.log10(good_data['DELTACHI2'])
print(len(x_good))
```

```python
all_number = []
good_number = []
for i_bin in range(0,len(log_delta_bin['average_bin'])):
    search = np.where((x_all>=log_delta_bin['bin'][i_bin]) & (x_all<log_delta_bin['bin'][i_bin+1]))
    all_number.append(len(search[0]))
    
    search = np.where((x_good>=log_delta_bin['bin'][i_bin]) & (x_good<log_delta_bin['bin'][i_bin+1]))
    good_number.append(len(search[0]))
    
print(all_number)
print(good_number)
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.axhline(1.,ls='--')
plt.plot(log_delta_bin['average_bin'],np.array(good_number)*1.0/all_number)
for i_bin in range(0,len(log_delta_bin['average_bin'])):
    plt.text(log_delta_bin['average_bin'][i_bin],(np.array(good_number)*1.0/all_number)[i_bin]-0.08,str(all_number[i_bin]))
    plt.text(log_delta_bin['average_bin'][i_bin],(np.array(good_number)*1.0/all_number)[i_bin]-0.04,str(good_number[i_bin]))

plt.ylim(0,1.1)
plt.subplot(1,2,2)
plt.axhline(1.,ls='--')
plt.plot(log_delta_bin['average_bin'],np.array(good_number)*1.0/all_number)
for i_bin in range(8,len(log_delta_bin['average_bin'])):
    plt.text(log_delta_bin['average_bin'][i_bin],(np.array(good_number)*1.0/all_number)[i_bin]-0.08,str(all_number[i_bin]))
    plt.text(log_delta_bin['average_bin'][i_bin],(np.array(good_number)*1.0/all_number)[i_bin]-0.04,str(good_number[i_bin]))

plt.xlim(2,5)
plt.ylim(0.6,1.05)
```
